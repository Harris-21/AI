# My Question

***What is the average Hp of all the pokemons?*** 

# My Analysis

# My Answer

***Using the three methods down below I've figrured out that the average HP of all the pokemons is 69.26.***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_df = pd.read_csv('pokemon.csv')
```

```python
average_hp = pokemon_df['HP'].mean()
```

```python
print(f"The average HP of the Pokémon is: {average_hp:.2f}")
```

```python
median_hp = pokemon_df['HP'].median()
```

```python
print(f"The median HP of the Pokémon is: {median_hp}")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(pokemon_df['HP'], bins=20, edgecolor='black', alpha=0.7)
plt.axvline(average_hp, color='red', linestyle='dashed', linewidth=1, label=f'Average HP: {average_hp:.2f}')
plt.title('Distribution of HP across Pokémon')
plt.xlabel('HP')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
