```python
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce')
```

```python
df = df.dropna(subset=['in_shazam_charts'])
```

```python
five_number_summary = df['in_shazam_charts'].describe(percentiles=[.25, .5, .75])
print(five_number_summary)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['in_shazam_charts'])
plt.title('Box Plot of in_shazam_charts')
plt.xlabel('Shazam Charts')
plt.show()
```

```python
df["streams"] = pd.to_numeric(df["streams"], errors="coerce")
df_solo = df[df['artist_count'] == 1]
df_collaboration = df[df['artist_count'] >= 1]

sns.histplot(data = df_collaboration, x = "streams", bins = 50, color = "Orange")
sns.histplot(data = df_solo, x = "streams", bins = 50, color = "Blue")


```
